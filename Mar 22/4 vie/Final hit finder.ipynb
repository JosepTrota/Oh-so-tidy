### Some setup functions...

```python
# to get genres for any string input

def getgenre(inputstring):
    result_ex = sp.search(str(inputstring))
    track_ex = result_ex['tracks']['items'][0]

    artist_ex = sp.artist(track_ex["artists"][0]["external_urls"]["spotify"])
    return(str(artist_ex["genres"]))

    album_ex = sp.album(track_ex["album"]["external_urls"]["spotify"])
    return(album_ex["genres"])
    return(album_ex["release_date"])

# to get song features for any string input
    
def getsongfeatures(inputstring):
    result = sp.search(str(inputstring))
    track = result['tracks']['items'][0]
    song_id = result['tracks']['items'][0]['id']
    return (sp.audio_features(str(song_id)))
```

```python
# to get all tracks in a playlist

tracks_ = pd.DataFrame()
def get_playlist_tracks(username,playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id)
    tracks_ = results['items']
    while results['next']:
        results = sp.next(results)
        tracks_.extend(results['items'])
    return tracks_
```

```python
# to display the songs

def spotify_display(song_name_id):
        display(IFrame(src=f"https://open.spotify.com/embed/track/{song_name_id}",
                        width="320",
                        height="80",
                        frameborder="0",
                        allowtransparency="true",
                        allow="encrypted-media",))
```

```python
# our recommender ;)

def recommend(userinput):
    newsongdata = getsongfeatures(str(userinput))[0]
    row_to_input = {}
    for column in huge_df:
        if str(column) in newsongdata.keys():
            row_to_input[str(column)]=newsongdata[str(column)]
        if str(column) in getgenre(str(userinput)):
            row_to_input[str(column)]=1
        else:
            if row_to_input[str(column)] == None:
                row_to_input[str(column)]=0       
    huge_df.append(row_to_input, ignore_index=True)
    scaler.fit(huge_df)
    huge_scaled=scaler.transform(huge_df)
    huge_KM_14=K_Means_14.fit(huge_scaled_df)
    huge_scaled_df['cluster_no']=huge_KM_14.labels_
    targetcluster = huge_scaled_df['cluster_no'][len(mydict_id)+1]
    huge_scaled_df.drop(huge_scaled_df.index[len(mydict_id)+1])
    print(targetcluster)
    rec = huge_scaled_df.sample()
    while str(rec['cluster_no']) != str(targetcluster):
        rec = huge_scaled_df.sample()
    print("This is your recomendation.")
    print("Now you are a tiny bit closer to enlightenment.")
    print("You're welcome ;)")
    spotify_display(str(rec['id'][0]))
    
```
